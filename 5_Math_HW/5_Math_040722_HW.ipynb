Now, let's plot the mean and the median on a plot. What we can do is create a vertical line at the mean and the median (preferably with different colors so we can distinguish the two). Below is an example of placing a vertical line on a plot.

```python
def normal_plot(xvalues, mu=0, sigma=1, cdf = False, **kwargs):
  '''
  Plots normal distribution probability distribution function or cumulative distribution function
  '''
  x = xvalues
  if cdf:
    y = stats.norm.cdf(x, mu, sigma)
  else:
    y = stats.norm.pdf(x, mu, sigma)
  plt.plot(x, y, **kwargs)

xvalues = np.linspace(-3, 12, 1000) # Try playing around with the first 2 numbers

# Try playing around with the values of the mean and standard deviation. Make sure the standard deviation is greater than 0!
mean = 3
standard_deviation = 4 

normal_plot(xvalues, mu = mean, sigma = standard_deviation) 
# normal_plot(xvalues, cdf = True)

# Plot a line at the mean
plt.axvline(x = mean, color = "red")
```
