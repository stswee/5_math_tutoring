```python
# Import libraries
import numpy as np
import random
import matplotlib.pyplot as plt
from statistics import stdev
```

```python
# Data
x =[3, 5, 7, 12, 5, 3, 4, 8, 10, 11]

# Values to calculate
mean = np.mean(x)
median = np.median(x)
mode = max(set(x), key=x.count) # Only gets the smallest mode
range = max(x) - min(x)

print(mean)
print(median)
print(mode)
print(range)
```

```python
# Low = # <- Lowest number
# High = # <- Highest number plus 1
# Size = # <- Number of data points

y = np.random.randint(low = 1, high = 11, size = 100) # Get numbers from 1 to 11-1=10
y = y.tolist()
print(y)

# Values to calculate
mean = np.mean(y)
median = np.median(y)
mode = max(set(y), key=y.count) # Only gets the smallest mode
range = max(y) - min(y)

print(mean)
print(median)
print(mode)
print(range)
```

```python
plt.hist(y)
```

The histogram above is easy to make, but difficult to read. 

```python
plt.hist(y, rwidth = 0.7, color = "red")
```

The histogram above is easier to read. 

```python
stdev(y)
```

Question: If I change only the maximum of the code below, will the standard deviation be higher or lower?

```python
a = np.random.randint(low = 1, high = 11, size = 100)
print(stdev(a))
b = np.random.randint(low = 1, high = 51, size = 100)
print(stdev(b))
```

Question: If I increase the size in the code, what happens to the standard deviation? 

```python
a = np.random.randint(low = 1, high = 11, size = 10)
print(stdev(a))
b = np.random.randint(low = 1, high = 11, size = 1000)
print(stdev(b))
```

If I have a large enough sample, I approach the "true" standard deviation. 

Take-home message: With a "small" sample size, the standard deviation is more difficult to predict.
