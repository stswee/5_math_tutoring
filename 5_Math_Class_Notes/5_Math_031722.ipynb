```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

# Mean, Median, Mode, and Range

```python
# Data
x = [3, 2, 5, 2, 5, 6, 2, 3, 6, 3, 9, 0, 2, 3, 5, 5, 3, 8, 4, 3, 7, 2, 9] 

mean = np.mean(x)
median = np.median(x)
mode = max(set(x), key=x.count)
range = max(x) - min(x)
print(mean)
print(median)
print(mode)
print(range)
```

# Histograms

```python
# Data
x = [3, 2, 5, 2, 5, 6, 2, 3, 6, 3, 9, 0, 2, 3, 5, 5, 3, 8, 4, 3, 7, 2, 9] 

# Plotting
plt.hist(x) # Default
```

```python
# Plotting and changing the number of bins
plt.hist(x, bins = 5)
```

Q: What happens when the number of bins increases? What happens when the number of bins decreases?

A: When the number of bins increases, the bin width gets smaller and the bars become more separate. When the number of bins decreases, the bin width gets bigger and the bars merge with each other.

```python
plt.hist(x, bins = 20)
```

# Boxplots

```python
plt.boxplot(x)
```

Minimum (0th percentile): Bottom line

Lower quartile (25th percentile): Bottom of the box

Median (50th percetile): Orange line

Upper quartile (75th percentile): Top of the box

Maximum (100th percentile): Top line

Q: Based on the image above, is the median closer or further from the lower quartile or the upper quartile?

A: Since the orange line is closer to the bottom of the box, then the median is closer to the lower quartile.

5 Number Summary

```python
quartiles = np.percentile(x, [0, 25, 50, 75, 100])
print(quartiles)
```

Q: What do each of the numbers (0, 25, 50, 75, and 100) mean?

A: They represent the percentiles in the data. In this case, quartiles want the percentiles to be 25 units apart.

```python
test = np.percentile(x, [10, 76])
print(test)
```

Q: What percentiles does test give?

A: Test gives the 10th and 76th percentiles of the data.

# Outliers

```python
# Data
y = [3, 2, 5, 2, 5, 6, 2, 3, 6, 3, 9, 0, 2, 3, 5, 5, 3, 8, 4, 3, 7, 2, 9, 50] # 25 is the outlier
```

```python
mean = np.mean(x)
median = np.median(x)
mode = max(set(x), key=x.count)
range = max(x) - min(x)
print(mean)
print(median)
print(mode)
print(range)
```

```python
mean = np.mean(y)
median = np.median(y)
mode = max(set(y), key=y.count)
range = max(y) - min(y)
print(mean)
print(median)
print(mode)
print(range)
```

Q: In general, which do you think is/are changed the most because of the outlier: mean, median, mode, or range?

A: The mean and range change a lot with the outlier. The median and mode do not change a lot with the outlier

```python
plt.boxplot(y)
```
